```python
class MyCallableClass:
    def __init__(self, message):
        self.message = message
        
    def __call__(self, name): # this special method is also called as "Dunder Call" method
        print(f"Hello {name}, {self.message}")
        
obj = MyCallableClass("Welcome to the Party!")
obj('Jaimin') # when we use parentheses with an object, python implicitly calls __call__ method!
obj.__call__("Jariwala")
```
