# Do following imports

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import random

import warnings
warnings.filterwarnings(action='ignore')
```

# Generate Fake Data

```python
x = torch.linspace(-5, 5, 50, requires_grad=True)
x_np = x.data.numpy() # numpy array for plotting
```

# Popular activation functions

```python
y_relu = F.relu(x).data.numpy() # relu() function takes in a tensor and not numpy.ndarray, hence provide 'x' and not 'x_np'
y_sigmoid = F.sigmoid(x).data.numpy()
y_tanh = F.tanh(x).data.numpy()
y_softplus = F.softplus(x).data.numpy()
y_leaky_relu = F.leaky_relu(x).data.numpy()
y_elu = F.elu(x).data.numpy()

# softmax function is used for multiclass classification problem
y_softmax = F.softmax(x).data.numpy() # softmax is a special type of activaiion function, it is about probability, and will make the sum as 1.
y_softmin = F.softmin(x).data.numpy()
```

# Activation function plot from above data

```python
cmap = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f']
color_iterator = iter(cmap)

y_list = [y_relu, y_sigmoid, y_tanh, y_softplus, y_leaky_relu, y_elu, y_softmax, y_softmin]
y_list_iterator = iter(y_list)

y_labels = ['relu', 'sigmoid', 'tanh', 'softplus', 'leaky_relu', 'elu', 'softmax', 'softmin']

plt.figure(figsize=(15,6), dpi=200)
for i in range(8):
    plt.subplot(2,4,i+1)
    plt.scatter(x_np, next(y_list_iterator),
                c=next(color_iterator),
                label=y_labels[i],
                marker='o',
                linewidths=1,
               edgecolors='black')
    plt.grid(alpha=0.7)
    plt.legend(loc='best')

plt.show()
```

# From above subplots, carefully look at the y_limits, whose range changes from 1 activation funciton to another!
